```python
import pickle
import pandas as pd
```

```python
ItemNewAll = pd.read_pickle('ItemNewAll.pkl')
ItemNewSpecial = pd.read_pickle('ItemNewSpecial.pkl')
Bestseller = pd.read_pickle('Bestseller.pkl')
BlogBest = pd.read_pickle('BlogBest.pkl')
```

```python
BlogBest.columns = ['title', 'author', 'pubDate', 'description', 'isbn13', 'categoryID',
       'categoryName', 'salesPoint', 'adult', 'rank', 'fullDes', 'fullDes2',
       'story', 'reviewrank', 'reviewtitle', 'review_w']
```

```python
BlogBest.head()
```

```python
data = pd.concat([ItemNewAll, ItemNewSpecial, Bestseller, BlogBest], ignore_index=True)
data = data.drop_duplicates()
data
```
